```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import numpy as np

# Load Excel file
file_path = Path(r"E:\USA_PD_2024\Analysis\ppr6\COP\ML\Feature_Selection_2\2.SFS\feature_importance_outputs\Best\SVM.xlsx")
df = pd.read_excel(file_path, sheet_name="SVM")

# Sort by mean importance
df_sorted = df.sort_values(by="Importance_mean", ascending=True)

# Clip lower error bar at 0 (no negative side)
means = df_sorted["Importance_mean"].values
errors = df_sorted["Importance_sd"].values
lower_errors = np.minimum(errors, means)  # Clip left error so it doesn't go below 0

# Create asymmetric error bars: [[lower], [upper]]
asymmetric_errors = [lower_errors, errors]

# Plotting
plt.figure(figsize=(10, 6))
plt.barh(
    df_sorted["Feature"],
    means,
    xerr=asymmetric_errors,
    capsize=5,
    color='steelblue'
)
plt.xlabel("Mean Feature Importance")
plt.title("Feature Importance (SVM) with SD")
plt.tight_layout()

# Save high-resolution PNG
output_path = file_path.with_name("svm_feature_importance_clipped.png")
plt.savefig(output_path, dpi=600, bbox_inches='tight')
print(f"Saved clipped plot to: {output_path}")
```
